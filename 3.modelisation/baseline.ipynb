# BASELINE MODELE 

### Import des données

```python
import pandas as pd

df = pd.read_csv("../6.Data/kaggle_b2_fraud_train_v3.csv")
df_test = pd.read_csv("../6.Data/kaggle_b2_fraud_test_v3.csv")
df.head()
```

### Creation du Baseline modèles (Majority Class)

```python
from sklearn.dummy import DummyClassifier
X_train = df.drop(columns=['target_is_fraud'])
y_train = df['target_is_fraud']

X_test = df_test


# DummyClassifier avec stratégie 'most_frequent'
baseline = DummyClassifier(strategy='most_frequent')
baseline.fit(X_train, y_train)
y_pred = baseline.predict(X_test)
```

### Sauvegarde du modèle

```python
submission = pd.DataFrame({
    'customer_id': df_test['customer_id'],     
    'target_is_fraud': y_pred      
})
submission.to_csv("../6.Data/baseline_submission.csv", index=False)
```
